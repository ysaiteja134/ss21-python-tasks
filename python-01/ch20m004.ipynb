## Problem 1 **Solution**

```python
def lagrange_polynomial(datalist, xvalue):
    x=[]
    y=[]
    yvalue=0
    for i in datalist:
        x.append(datalist[i][0])
        y.append(datalist[i][1])
    l = len(datalist)
    for i in range(l):
        term = 1
        for j in range(l):
            if i!=j:
                term = term*(xvalue-x[j])/(x[i]-x[j])
        yvalue += term*y[i]
    return yvalue
        
```
